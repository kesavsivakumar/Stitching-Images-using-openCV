```
import cv2
import matplotlib.pyplot as plt
import glob
```

```
images=[]

for path in glob.glob('/content/drive/My Drive/Untitled folder/*.jpg'):
  print(path)
  img=cv2.imread(path)
  img=cv2.resize(img,(256,256))
  images.append(img)
```

```
len(images)
```

```
stitcher = cv2.Stitcher.create(mode = 1)
```

```
x=stitcher.stitch(images)
```
